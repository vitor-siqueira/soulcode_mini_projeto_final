# Traduzindo o Schema para Portugu√™s-BR

```python
df.rename(columns={'Year_Birth': 'ano_de_nascimento',
                   'Education': 'graduacao', #grau de instru√ß√£o do cliente
                   'Marital_Status': 'estado_civil',
                   'Income': 'renda_anual', #renda familiar anual do cliente
                   'Kidhome': 'criancas_em_casa', #n√∫mero de crian√ßas pequenas na fam√≠lia do cliente
                   'Teenhome': 'adolescentes_em_casa', #n√∫mero de adolescentes na fam√≠lia do cliente
                   'Dt_Customer': 'data_cliente', #data de inscri√ß√£o do cliente na empresa
                   'Recency': 'recencia', #n√∫mero de dias desde a √∫ltima compra
                   'MntWines': 'qts_vinhos', #valor gasto em produtos v√≠nicos nos √∫ltimos 2 anos
                   'MntFruits': 'qts_frutas', #montante gasto em produtos de frutas nos √∫ltimos 2 anos
                   'MntMeatProducts': 'qts_produtos_carne', #quantia gasta em produtos de carne nos √∫ltimos 2 anos
                   'MntFishProducts': 'qts_produtos_peixe', #quantia gasta em produtos de peixe nos √∫ltimos 2 anos
                   'MntSweetProducts': 'qts_produtos_doces', #valor gasto em produtos doces nos √∫ltimos 2 anos
                   'MntGoldProds': 'qts_produtos_ouro', #valor gasto em produtos de ouro nos √∫ltimos 2 anos
                   'NumDealsPurchases': 'num_compras_desconto', #n√∫mero de compras feito com desconto
                   'NumWebPurchases': 'num_compras_web', #n√∫mero de compras feitas atrav√©s do site da empresa
                   'NumCatalogPurchases': 'num_compras_catalogo', #n√∫mero de compras feitas usando o cat√°logo
                   'NumStorePurchases': 'num_compras_loja', #n√∫mero de compras feitas diretamente nas lojas
                   'NumWebVisitsMonth': 'num_visitas_web_mes', #n√∫mero de visitas ao site da empresa no √∫ltimo m√™s
                   'Complain': 'reclamacoes', #1 se o cliente reclamou nos √∫ltimos 2 anos
                   'AcceptedCmp1': 'campanha1', #1 se o cliente aceitou a oferta na 1¬™ campanha, 0 caso contr√°rio
                   'AcceptedCmp2': 'campanha2', #1 se o cliente aceitou a oferta na 2¬™ campanha, 0 caso contr√°rio
                   'AcceptedCmp3': 'campanha3', #1 se o cliente aceitou a oferta na 3¬™ campanha, 0 caso contr√°rio
                   'AcceptedCmp4': 'campanha4', #1 se o cliente aceitou a oferta na 4¬™ campanha, 0 caso contr√°rio
                   'AcceptedCmp5': 'campanha5', #1 se o cliente aceitou a oferta na 5¬™ campanha, 0 caso contr√°rio
                   'Response': 'resposta_alvo' #1 se o cliente aceitou a oferta na √∫ltima campanha, 0 caso contr√°rio
                   }, inplace = True)
```

```python
df.drop(['Z_CostContact', 'Z_Revenue'], axis = 1, inplace = True)
```

# Deletando os anos de nascimento que n√£o fazem sentido com o dataframe

```python
df.drop(df.loc[df.ano_de_nascimento == 1893].index, inplace = True)
df.drop(df.loc[df.ano_de_nascimento == 1899].index, inplace = True)
df.drop(df.loc[df.ano_de_nascimento == 1900].index, inplace = True)
```

#Localizando e traduzindo os dados para Portugu√™s-BR

```python
#graduacao
df.loc[df.graduacao == "Graduation", ['graduacao']] = 'Ensino Superior'
df.loc[df.graduacao == "Master", ['graduacao']] = 'Mestrado'
df.loc[df.graduacao == "Basic", ['graduacao']] = 'Ensino M√©dio'
df.loc[df.graduacao == "2n Cycle", ['graduacao']] = 'Cursando Mestrado'

#estado_civil
df.loc[df.estado_civil == "Single", ['estado_civil']] = 'Solteiro(a)'
df.loc[df.estado_civil == "Together", ['estado_civil']] = 'Uni√£o Est√°vel'
df.loc[df.estado_civil == "Married", ['estado_civil']] = 'Casado(a)'
df.loc[df.estado_civil == "Divorced", ['estado_civil']] = 'Divorciado(a)'
df.loc[df.estado_civil == "Widow", ['estado_civil']] = 'Vi√∫vo(a)'
df.loc[df.estado_civil == "Alone", ['estado_civil']] = 'Sozinho(a)'
df.loc[df.estado_civil == "YOLO", ['estado_civil']] = 'Aproveitando a vida'

# Transformando o estado civil Absurd em NA
df.replace(['Absurd'], pd.NA, inplace = True)
```

#Deletando os NA somente das colunas informadas

```python
df.dropna(subset=['estado_civil'], inplace = True)
```

```python
df.dropna(subset=['renda_anual'], inplace = True)
```
