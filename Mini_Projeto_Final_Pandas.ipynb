# Traduzindo o Schema para Português-BR

```python
df.rename(columns={'Year_Birth': 'ano_de_nascimento',
                   'Education': 'graduacao', #grau de instrução do cliente
                   'Marital_Status': 'estado_civil',
                   'Income': 'renda_anual', #renda familiar anual do cliente
                   'Kidhome': 'criancas_em_casa', #número de crianças pequenas na família do cliente
                   'Teenhome': 'adolescentes_em_casa', #número de adolescentes na família do cliente
                   'Dt_Customer': 'data_cliente', #data de inscrição do cliente na empresa
                   'Recency': 'recencia', #número de dias desde a última compra
                   'MntWines': 'qts_vinhos', #valor gasto em produtos vínicos nos últimos 2 anos
                   'MntFruits': 'qts_frutas', #montante gasto em produtos de frutas nos últimos 2 anos
                   'MntMeatProducts': 'qts_produtos_carne', #quantia gasta em produtos de carne nos últimos 2 anos
                   'MntFishProducts': 'qts_produtos_peixe', #quantia gasta em produtos de peixe nos últimos 2 anos
                   'MntSweetProducts': 'qts_produtos_doces', #valor gasto em produtos doces nos últimos 2 anos
                   'MntGoldProds': 'qts_produtos_ouro', #valor gasto em produtos de ouro nos últimos 2 anos
                   'NumDealsPurchases': 'num_compras_desconto', #número de compras feito com desconto
                   'NumWebPurchases': 'num_compras_web', #número de compras feitas através do site da empresa
                   'NumCatalogPurchases': 'num_compras_catalogo', #número de compras feitas usando o catálogo
                   'NumStorePurchases': 'num_compras_loja', #número de compras feitas diretamente nas lojas
                   'NumWebVisitsMonth': 'num_visitas_web_mes', #número de visitas ao site da empresa no último mês
                   'Complain': 'reclamacoes', #1 se o cliente reclamou nos últimos 2 anos
                   'AcceptedCmp1': 'campanha1', #1 se o cliente aceitou a oferta na 1ª campanha, 0 caso contrário
                   'AcceptedCmp2': 'campanha2', #1 se o cliente aceitou a oferta na 2ª campanha, 0 caso contrário
                   'AcceptedCmp3': 'campanha3', #1 se o cliente aceitou a oferta na 3ª campanha, 0 caso contrário
                   'AcceptedCmp4': 'campanha4', #1 se o cliente aceitou a oferta na 4ª campanha, 0 caso contrário
                   'AcceptedCmp5': 'campanha5', #1 se o cliente aceitou a oferta na 5ª campanha, 0 caso contrário
                   'Response': 'resposta_alvo' #1 se o cliente aceitou a oferta na última campanha, 0 caso contrário
                   }, inplace = True)
```

# Deletando colunas desnecessárias

```python
df.drop(['Z_CostContact', 'Z_Revenue'], axis = 1, inplace = True)
```

# Deletando os anos de nascimento que não fazem sentido com o dataframe

```python
df.drop(df.loc[df.ano_de_nascimento == 1893].index, inplace = True)
df.drop(df.loc[df.ano_de_nascimento == 1899].index, inplace = True)
df.drop(df.loc[df.ano_de_nascimento == 1900].index, inplace = True)
```

#Localizando e traduzindo os dados para Português-BR

```python
#graduacao
df.loc[df.graduacao == "Graduation", ['graduacao']] = 'Ensino Superior'
df.loc[df.graduacao == "Master", ['graduacao']] = 'Mestrado'
df.loc[df.graduacao == "Basic", ['graduacao']] = 'Ensino Médio'
df.loc[df.graduacao == "2n Cycle", ['graduacao']] = 'Cursando Mestrado'

#estado_civil
df.loc[df.estado_civil == "Single", ['estado_civil']] = 'Solteiro(a)'
df.loc[df.estado_civil == "Together", ['estado_civil']] = 'União Estável'
df.loc[df.estado_civil == "Married", ['estado_civil']] = 'Casado(a)'
df.loc[df.estado_civil == "Divorced", ['estado_civil']] = 'Divorciado(a)'
df.loc[df.estado_civil == "Widow", ['estado_civil']] = 'Viúvo(a)'
df.loc[df.estado_civil == "Alone", ['estado_civil']] = 'Sozinho(a)'
df.loc[df.estado_civil == "YOLO", ['estado_civil']] = 'Aproveitando a vida'

# Transformando o estado civil Absurd em NA
df.replace(['Absurd'], pd.NA, inplace = True)
```

```python
df.head()
```

#Deletando os NA somente das colunas informadas

```python
df.dropna(subset=['estado_civil'], inplace = True)
```

```python
df.dropna(subset=['renda_anual'], inplace = True)
```

# Verificando se os dados <NA> foram deletados

```python
df.isna().sum()
```

```python
print(df.info())
```

Após as transformações realizadas, podemos verificar que as colunas foram todas traduzidas para o Português-BR, não há mais nenhum dado faltante após a realização da limpeza dos NA
